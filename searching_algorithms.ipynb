```python
#linear search
n=int(input("enter number of elemnts"))
arr=[]
for i in range(n):
  arr.append(int(input(f"ENter element {i+1}: ")))
key = int(input("Enter element to search:"))
found=False
for i in range(n):
  if arr[i]==key:
    print(f"Element {key} found at position {i+1}")
    found=True
    break
if not found:
    print(f"element {key} not found")
```

```python
#binary search
n=int(input("Enter no. of elements"))
arr=[]
for i in range(n):
  arr.append(int(input(f"Enter element {i+1}:")))
arr.sort()
print("\nSorted list",arr)
key=int(input("Enter element to search:"))
low=0
high=n-1
found=False
while low<=high:
  mid=(low+high)//2
  if arr[mid]==key:
    print(f"Element {key} found position {mid+1}")
    found=True
    break
  elif key<arr[mid]:
    high=mid-1
  else:
    low=mid+1
if not found:
  print("Element not found4")
```
