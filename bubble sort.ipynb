```python
#bubble sort
n=int(input("Enter no. of elemnts"))
arr=[]
for i in range(n):
  arr.append(int(input(f"Enter element {i+1}: ")))
print("\n Original ARRay :", arr)
for i in range(n-1):
  for j in range(n-1-i):
    if arr[j]>arr[j+1]:
      arr[j],arr[j+1]=arr[j+1],arr[j]
print("Sorted Array: ",arr)
```
