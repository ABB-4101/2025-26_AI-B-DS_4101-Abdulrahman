```python
#linear search
n=int(input("enter number of elemnts"))
arr=[]
for i in range(n):
  arr.append(int(input(f"ENter element {i+1}: ")))
key = int(input("Enter element to search:"))
found=False
for i in range(n):
  if arr[i]==key:
    print(f"Element {key} found at position {i+1}")
    found=True
    break
if not found:
    print(f"element {key} not found")
```
